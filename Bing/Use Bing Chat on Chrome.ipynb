In order to use Bing chat on Chrome, you'll will need to trick Bing into thinking that you're running it from an Edge browser.

In order to do this, first open up Chrome and go to [Microsoft Bing](https://www.bing.com/) and then press **F12**.

Under the **Network** tab go to the **Network conditions** found on the bottom of the screen. Make sure that the **User browser default** is unselected and use the drop-down menu to select a Microsoft Edge browser.



Before you close the developer window, select Chat from the browser and once it enters the chat portal, you can then close it.



Enter your prompt and enjoy Bing on Chrome!
